```python
# Select only the numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=np.number)
numeric_df.corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
# Top 10 Most Populous Countries
top_10 = df.sort_values(by='2022 Population', ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x='Country', y='2022 Population', data=top_10)
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Most Populous Countries')
plt.xlabel('Country')
plt.ylabel('2022 Population')
plt.show()
```

```python
# Distribution of 2022 Population
plt.figure(figsize=(8, 4))
sns.histplot(df['2022 Population'], bins=30, kde=True)
plt.title('Distribution of 2022 Population')
plt.xlabel('2022 Population')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot of population density vs. 2022 Population
plt.figure(figsize=(8, 4))
sns.scatterplot(x='Density (per km²)', y='2022 Population', data=df)
plt.title('Population Density vs. 2022 Population')
plt.xlabel('Population Density (per km²)')
plt.ylabel('2022 Population')
plt.show()
```

```python
# Comparing population across continents using a boxplot
plt.figure(figsize=(12,6))
sns.boxplot(x='Continent', y='2022 Population', data=df)
plt.xticks(rotation=45, ha='right')
plt.title('Population Distribution Across Continents')
plt.show()
```

```python
# Pie chart of the population distribution across continents in 2022.
continent_population = df.groupby('Continent')['2022 Population'].sum()
plt.figure(figsize=(8, 4))
plt.pie(continent_population, labels=continent_population.index, autopct='%1.1f%%', startangle=90)
plt.title('Population Distribution Across Continents in 2022')
plt.show()
```
